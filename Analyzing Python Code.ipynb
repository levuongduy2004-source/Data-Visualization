```python
import pandas as pd
import plotly.express as px
import plotly.io as pio
```

```python
df = pd.read_csv('shopping_behavior_updated.csv')
df.info()
```

```python
df['Customer ID'] = df['Customer ID'].astype('object')
```

```python
df.to_csv('shopping_behavior_updated.csv')
```

```python
df.info()
```

```python
#1 Bar Chart – Total Purchase Amount by Category
bar_fig = px.bar(
    df.groupby('Category', as_index=False)['Purchase Amount (USD)'].sum(),
    x='Category', y='Purchase Amount (USD)',
    title='Total Purchase Amount by Category',
    color='Category', text_auto='.2s'
)

bar_fig.update_layout(xaxis_title='Category', yaxis_title='Total Purchase Amount (USD)', title_font_size=20)
bar_fig.show()
```

```python
#2 Box Plot – Distribution of Purchase Amount per Category
box_fig = px.box(
    df, x='Category', y='Purchase Amount (USD)', color='Category',
    title='Distribution of Purchase Amount per Category (Box Plot)'
)
box_fig.update_layout(title_font_size=20)
box_fig.show()
```

```python
#3 Scatter Plot – Relationship between Age and Purchase Amount
scatter_fig = px.scatter(
    df, x='Age', y='Purchase Amount (USD)', color='Category',
    hover_data=['Item Purchased', 'Season'],
    title='Relationship between Age and Purchase Amount by Category',
    trendline='ols'
)
scatter_fig.update_layout(title_font_size=20)
scatter_fig.show()
```

```python
#4 Treemap – Category > Item Purchased by Total Purchase Amount
tree_fig = px.treemap(
    df, path=['Category', 'Item Purchased'], values='Purchase Amount (USD)',
    title='Revenue Structure by Category and Item'
)
tree_fig.update_layout(title_font_size=20)
tree_fig.show()
```

```python
#5 Donut Chart – Purchase Amount by Season
season_sum = df.groupby('Season', as_index=False)['Purchase Amount (USD)'].sum()
donut_fig = px.pie(
    season_sum, names='Season', values='Purchase Amount (USD)',
    hole=0.5, title='Seasonal Purchase Distribution'
)
donut_fig.update_traces(textinfo='percent+label')
donut_fig.show()
```

```python
#6 Histogram – Review Rating Distribution
hist_fig = px.histogram(
    df, x='Review Rating', nbins=10, color_discrete_sequence=['#1f77b4'],
    title='Distribution of Customer Review Ratings'
)
hist_fig.update_layout(xaxis_title='Rating', yaxis_title='Count', title_font_size=20)
hist_fig.show()
```
